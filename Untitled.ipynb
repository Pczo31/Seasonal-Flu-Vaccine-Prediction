```python
imp_train = SimpleImputer(missing_values = np.nan, strategy = 'median')
imp_train
```

```python
numeric_columns = [col for col in df3.columns if df3[col].dtype == 'float64']
```

```python
numeric_training_data = df3.loc[:,numeric_columns]

```

```python
corr_matrix = numeric_training_data.corr()
```

```python
np.max(np.tril(corr_matrix, k = -1))
```

```python
mask = np.triu(corr_matrix)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

cmap = sns.diverging_palette(500, 8, as_cmap=True)

sns.heatmap(corr_matrix, mask = mask, cmap = cmap, center = 0,
            square = True, linewidths = .5, cbar_kws = {"shrink": .5})
plt.show()

```

```python
train_numeric_imputed = pd.DataFrame(imp_train.fit_transform(numeric_training_data),columns = numeric_training_data.columns)
train_numeric_imputed
```

```python
h1n1_model = LogisticRegression()
seasonal_model = LogisticRegression()
```

```python
h1n1_model.fit(train_numeric_imputed, df4['h1n1_vaccine'])
seasonal_model.fit(train_numeric_imputed, df4['seasonal_vaccine'])
```

```python
h1n1_importance = h1n1_model.coef_[0]
h1n1_importance
```

```python
for index, (col_name,importance) in enumerate(zip(train_numeric_imputed.columns, h1n1_importance)):
    print(f"Importance for the '{col_name}' column (Column # {index}) : {importance}")

plt.bar([x for x in range(len(h1n1_importance))], h1n1_importance)
plt.show()
```
